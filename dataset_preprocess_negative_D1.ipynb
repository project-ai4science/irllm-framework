### Data format
The data format is a .pkl file (pickle) containing a dataframe with at least these columns:
- id, a_text, b_id, b_text, c_id, c_text, research_type, y_true

```python
import pandas as pd

all_papers = pd.read_csv("./data/temp/papers_1subj_2025.csv")
print(all_papers.shape, "->", all_papers.columns)
```

```python
data = all_papers.sample(1000, random_state=42).reset_index(drop=True)

print(data.shape, "->", data.columns)
```

```python
# Getting a single string for "paper B" and a single string for "paper C"

def concatTitleAbstract(obj):
    if obj.get('abstract', None):
        return f"{obj['title']}; {obj['abstract']}"
    return obj['title']

# data["id"] = None
data["a_text"] = None
data["b_id"] = None
data["b_text"] = None
data["c_id"] = None
data["c_text"] = None

for i, row in data.iterrows():
    a_text = concatTitleAbstract(row)
    b_text = ""#concatTitleAbstract(row[0])
    c_text = ""#concatTitleAbstract(row[1])
    
    data.loc[i, "a_text"] = a_text
    data.loc[i, "b_id"] = ""
    data.loc[i, "b_text"] = b_text
    data.loc[i, "c_id"] = ""
    data.loc[i, "c_text"] = c_text

data["y_true"] = False
data["research_type"] = None

print(data.shape, "->", data.columns)
```

```python
data.to_pickle("./data/data_D1.pkl")
data = pd.read_pickle("./data/data_D1.pkl")

data
```
